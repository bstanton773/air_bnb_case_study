```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighbourhood groups / which shows up the most
air_bnb.groupby('neighbourhood_group').count()['id'].sort_values(ascending=False)
```

```python
# Are private rooms the most popular in Manhattan?
manhattan_bnb = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan']
manhattan_bnb.groupby('room_type').count()['id'].sort_values(ascending=False)
```

```python
air_bnb.head()
```

```python
# Who are the top 10 busiest hosts based on number of reviews per month?
air_bnb.groupby(['host_id', 'host_name']).sum()['reviews_per_month'].nlargest(10)
```

```python
# Which neighborhood group has the highest avg price and plot the average prices in a bar chart
avg_prices = air_bnb.groupby('neighbourhood_group').mean()['price'].nlargest()
avg_prices
```

```python
plt.figure(figsize=(20,10))
plt.bar(avg_prices.index, avg_prices, color=['r', 'g', 'b', 'k', 'y'])
plt.title('Average Price per Burough in New York', fontname='Times New Roman', fontsize=32)
plt.grid(True)
plt.ylabel('Average Price', fontname='Times New Roman', fontsize=24)
plt.xlabel('Burough', fontname='Times New Roman', fontsize=24)
plt.show()
```

```python
# What is the most expensive neighbourhood in each neighbourhood_group
avg_group = air_bnb.groupby(['neighbourhood_group', 'neighbourhood']).mean()
avg_group
```

```python
bronx_highest = avg_group.loc['Bronx'].nlargest(1, 'price')
bronx_highest
```

```python
brooklyn_highest = avg_group.loc['Brooklyn'].nlargest(1, 'price')
brooklyn_highest
```

```python
manhattan_highest = avg_group.loc['Manhattan'].nlargest(1, 'price')
manhattan_highest
```

```python
queens_highest = avg_group.loc['Queens'].nlargest(1, 'price')
queens_highest
```

```python
staten_island_highest = avg_group.loc['Staten Island'].nlargest(1, 'price')
staten_island_highest
```

```python
# Which host has the highest minimum rental? *minimal rental is based on price and minimum nights
air_bnb['minimum_rental'] = air_bnb['minimum_nights'] * air_bnb['price']
air_bnb.nlargest(5, 'minimum_rental')
```

```python
air_bnb[['host_name','minimum_nights','price']].nlargest(10, ['price'], keep='last').sort_values('minimum_nights').head(1)
```

```python
# Who has no (zero) availability with a review count of 100 or more
air_bnb[(air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] >= 100)]
```

```python
# If each host rents all of their properties for the night, which 5 make the most that night?
air_bnb.groupby(['host_id', 'host_name']).sum().nlargest(5, 'price')
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')].groupby('host_id').max()
```

```python
air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')].sort_values('last_review')
```

```python
# bar chart showing the number of room types in the Bronx
bronx_bnb = air_bnb[air_bnb['neighbourhood_group'] == 'Bronx']
room_counts = bronx_bnb.groupby('room_type').count()['id']
room_counts
```

```python
with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(figsize=(15,10))
    ax.bar(room_counts.index, room_counts, color=['r', 'g', 'k'])
    plt.title('Number of Rooms per Room Type in the Bronx', fontsize=24)
    plt.xlabel('Room Type')
    plt.ylabel('Number of Rooms')
    plt.show()
```
